```python
from sklearn.datasets import load_iris
iris=load_iris() #() runs function and returns Bunch object as iris
print(iris)
```

```python
np.mean(iris.data) #mean of all 600 scalars, but want mean of each column separately
np.mean(iris.data,axis=1) #takes mean of each row->not meaningful here, gives 150 numbers
np.mean(iris.data,axis=0) #takes mean of each column to give mean of 1 feature across 150 samples, for 4 features->4 numbers
mean_vector=np.mean(iris.data,axis=0) #assign to variable
print(mean_vector)
```

```python
x = iris.data #good notation makes things cleaner later
print(x[0:5])
mean_fvector = np.mean(x, axis=0) #avg columns of x, assigned to variable
print(mean_fvector)
x_c = x - mean_fvector #subtracts mean vector from each row, does so by "broadcasting" automatic in NumPy gives X centered data, mean as 0,0,0,0 new origin
print(x_c[0:5]) #just show first 5 so doesn't take up entire screen
```

```python
#x_cTx_c gives self adjoint operator as covariance feature matrix (when divided by n-1), rotate coordinate systems with new axes that better explain spread/variance of data
#new axes may be linear combination of previous bases-septal length, width, etc..

# X: map from feature to sample space: standard metric on sample space, using x-pullback metric to feature space, geometry of feature space induced by how features map to sample
#gF:<v,w> metric on feature space defines inner product of v,w elements of F equal to gS:<Xv,Xw> metric on sample space with inner products of X transforming v and w
#Algebra of inner product: (Xv)T(Xw)= vTXTXw this is inner product on feature space, induced by the sample space (which has I=inner product)
#this is the pullback metric and warps the feature space according to how its embedded in sample space, gives us an ellipsoid from the unit circle with PC axes
```

```python
n=x_c.shape[0] #x_c is centered data, its shape is 150,4 and we only look at first number to get number of samples
C_f = (x_c.T @ x_c) / (n-1) #define covariance feature matrix as x_centered Transpose matrix muplitied @ by x_centered, then divided by n-1
print("feature covariance matrix: \n" , C_f) #\n new line, clean
#diagonal is variance, off diagonal is covariance:how features vary together. Eigenvectors are principal components=new axis directions, eigenvalue is its variance
```

```python
#x_c has a data cloud centered at the origin of feature space
#variance: project all points onto 1 axis, the more they spread from the origin the greater variance (deviation from mean squared)-variance is self covariance
#covraince: measurse how two features vary together across samples (not data points)
#covariance matrix: dot product of centered feature vectors
#xTx pullback matrix from sample space to feature space, due to feature space's embedding in sample space= data cloud
#xxT pushforward of metric from feature space to sample space: tells us how samples are similar using the feature metric= how samples cluster
#xTx covar matrix defines how feature vectors v,w are measured according to image under X
```

```python
#for symmetric/hermitian matrices can use np.linalg.eigh, h for hermitian
eigen_decomp = np.linalg.eigh(C_f)
eigenvalues = eigen_decomp[0]
eigenvectors = eigen_decomp[1]
print(eigenvalues) #4 eigenvalues
print(eigenvectors) # 4 x 4 matrix
```

```python
#np.argsort() to rearrange from ascending to decreasing, maximum variance is what matters for PCA
np.argsort(eigenvalues) #this does not sort values, but rather states the index that comes first (with smallest value)
#slicing array[start:stop:step] because step=-1, goes backward thus start if step<0 then start=-1 (which is end of the index)
d_order= np.argsort(eigenvalues)[::-1]
print(d_order) #gives index in reverse orer
sorted_eigenvalues = eigenvalues[d_order]
print(sorted_eigenvalues)

#eigenvectors is a matrix, with columns of eigenvectors by NumPy slicing array[rows,cols] so [:, d_order], takes all the rows of each column, then rearrange columns by d_order
sorted_eigenvectors = eigenvectors[:, d_order]
print(sorted_eigenvectors)
```

```python
#project centered data on new axes=directions of max variance, new bases, gives coordinates in reduced feature space=PCA space
#can use smaller space-dim reduction, as variance mostly captured by few components
k = 1  # however many components you want-> pick k s.t. it encompasses 90-95% variance, i.e. eigen/sum of eigen x 100> 95, in this case can just use 1
principal_components = sorted_eigenvectors[:, :k]  # shape (n_features, k)
print(principal_components)
#project each sample onto subspace spanned by principal components
x_pca= x_c @ principal_components #shape (n_samples, k)
#in this case, project centered data onto 1D- the principal component axis, compresses 4D while preserving maximum variance
#in general, PCA finds lowerD submanifold that approximates data manifodl well-preserving maximal Fisher Information(variance)
#isntead of measuring sensitivity of probab distrubtions wrt parameters, high variance as parameters w/ more influence on data
```

```python
import matplotlib.pyplot as plt
import numpy as np

#shape is (150, 1):150 arrays all holding one number, flatten to 1D array: one array with 150 numbers: Easier to work with whe plotting on one line
pc1_coords = x_pca.flatten()
plt.figure(figsize=(10, 2))  # wide and short figure for a horizontal line, this defines the canvas ideal for 1D line


for species in set(iris.target): #set removes repeats-only unique class labels 0,1,2
# iris.target == species gives boolean array for a label e.g. iris.target == 0 returns [T,T,...,F,F]
#np.where returns the indices where this is true, gives positions of samples that belong to that class. In this case for 0, [0,...,49]
    indices = np.where(iris.target == species)
#pc1_coords[indices] # selects data from indices:projected PCA values for that class
#iris.target has class labels, and iris.target_names maps them to array([setosa,versi,virg])

    # Scatter plot: x = PCA coordinate, y = small random jitter so points don't overlap
    plt.scatter(pc1_coords[indices], #selects PCA1D values for given label, plot along x-axis
                np.random.uniform(-0.05, 0.05, size=len(indices[0])), #spreads dots vertically, visually: random jitter so don't overlap
                label=iris.target_names[species], alpha=0.7) #alpha controls transparency to see overlapping dots
#loop repeats the plotting 3 times, once for each class

#plt.yticks([])  # hide y-axis ticks (since y has no meaning here)
plt.xlabel('First Principal Component Value')
plt.title('Iris Dataset Projected onto 1D PCA Axis')
plt.legend() #collects all labels from label in plt.scatter and gives legendbox
plt.show()
```

```python
# Define k=2 principal components and project data
k = 2
principal_components_2d = sorted_eigenvectors[:, :k]  # shape (n_features, 2)
print(principal_components_2d)

# Project centered data onto 2D PCA space
x_pca_2d = x_c @ principal_components_2d  # shape (n_samples, 2)

# Visualization of 2D PCA projection
import matplotlib.pyplot as plt
import numpy as np

pc1_coords = x_pca_2d[:, 0]  # First principal component (x-axis)
pc2_coords = x_pca_2d[:, 1]  # Second principal component (y-axis)

plt.figure(figsize=(8, 6))

for species in set(iris.target):
    indices = np.where(iris.target == species)
    plt.scatter(
        pc1_coords[indices],
        pc2_coords[indices],
        label=iris.target_names[species],
        alpha=0.7
    )

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('Iris Dataset Projected onto 2D PCA')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define k=3 principal components and project data
k = 3
principal_components_3d = sorted_eigenvectors[:, :k]  # shape (n_features, 3)
print(principal_components_3d)

# Project centered data onto 3D PCA space
x_pca_3d = x_c @ principal_components_3d  # shape (n_samples, 3)

# Visualization of 3D PCA projection
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # needed for 3D plotting
import numpy as np

pc1_coords = x_pca_3d[:, 0]  # PC1
pc2_coords = x_pca_3d[:, 1]  # PC2
pc3_coords = x_pca_3d[:, 2]  # PC3

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for species in set(iris.target):
    indices = np.where(iris.target == species)
    ax.scatter(
        pc1_coords[indices],
        pc2_coords[indices],
        pc3_coords[indices],
        label=iris.target_names[species],
        alpha=0.7
    )

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('Iris Dataset Projected onto 3D PCA')
ax.legend()
plt.show()
```
